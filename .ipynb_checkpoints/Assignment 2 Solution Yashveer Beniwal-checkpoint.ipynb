````python
import random
import pandas as pd
import numpy as np

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
````

````python
def plot_probs(x_, y_):
    trace = go.Scatter(
        x = x_,
        y = y_
    )

    layout = go.Layout(
        xaxis=dict(
            title='Number of Simulations'
        ),
        yaxis=dict(
            title='Probability'
        )
    )

    fig = go.Figure(data = [trace], layout = layout)
    return fig
````

# Question 1 

Twins. About 30% of human twins are identical, and the rest are fraternal. Identical twins are necessarily the same sex – half are males and the other half are females. One-quarter of fraternal twins are both male, one-quarter both female, and one-half are mixes: one male, one female. You have just become a parent of twins and are told they are both girls. Given this information, what is the probability that they are identical? 

# Answer 

<img src="https://github.com/ysvrbnwl/ysvrbnwl.github.io/blob/master/twins.jpg?raw=true" />

```bash
P(identical) = 0.30
    ├── P(males|identical) = 0.50
    ├── P(females|identical) = 0.50
    ├── P(mixes|identical) = 0
    
P(identical) = 0.70
    ├── P(males|identical) = 0.25
    ├── P(females|identical) = 0.25
    ├── P(mixes|identical) = 0.50
```   
```
P(identical|females) = P(identical & females) / P(females)
P(identical|females) = P(identical & females) / (P(identical & females) + P(fraternal & females))
P(identical|females) = (0.30 * 0.50) / (0.30 * 0.50 + 0.70 * 0.25)
P(identical|females) = 0.15 / 0.325 = 0.461
```
### Simulation

````python
prob_twins = [0.30, 0.70] # [identical, fraternal]
prob_identical = [0.50, 0.50] # [males, females]
prob_fraternal = [0.25, 0.25, 0.50] # [males, females, mixes]

identical_females = 0
females = 0
x_data = []
y_data = []

while females < 10000:
    twins = np.random.choice(['identical', 'fraternal'], p=prob_twins)
    if twins == 'fraternal':
        gender = np.random.choice(['males', 'females', 'mixes'], p=prob_fraternal)
        if gender == 'females':
            females += 1
    elif twins == 'identical':
        gender = np.random.choice(['males', 'females'], p=prob_identical)
        if gender == 'females':
            identical_females += 1
            females += 1
    if females != 0 and (females % 10) == 0:
        y_data.append(float(identical_females) / float(females))
        x_data.append(females)
    
        
# Plot the figure
fig = plot_probs(x_data, y_data)
iplot(fig)
````

# Question 2

Sickle cell anemia. Sickle cell anemia is a genetic blood disorder where red blood cells lose their flexibility and assume an abnormal, rigid, “sickle” shape, which results in a risk of various complications. If both parents are carriers of the disease, then a child has a 25% chance of having the disease, 50% chance of being a carrier, and 25% chance of neither having the disease nor being a carrier. If two parents who are carriers of the disease have 3 children, what is the probability that 

(a) two will have the disease? 

(b) none will have the disease? 

(c) at least one will neither have the disease nor be a carrier? 

(d) the first child with the disease will the be 3rd child? 

# Answer 


| `Symbol`| `Description`|`Probablity`|
|:-----:|-----| :----:|
| D | Disease | 0.25 |
| C | Carrier | 0.50|
| N | Neither having the disease nor being a carrier| 0.25 |



## a) two will have disease?
```
ans = P(DDC) + P(DCD) + P(CDD) +
      P(DDN) + P(DND) + P(NDD)
ans = 3 * (0.25 * 0.25 * 0.50) + 3 * (0.25 * 0.25 * 0.25)
ans = 0.140625
```
## b) none will have the disease?
```
ans = P(NNN) + P(NNC) + P(NCN) + P(CNN) +
      P(CCC) + P(NCC) + P(CNC) + P(CCN)
ans = (0.25 * 0.25 * 0.25) + 3 * (0.25 * 0.25 * 0.50) + 3 * (0.25 * 0.50 * 0.50) + (0.50 * 0.50 * 0.50) 
ans = 0.421875
```
## c) at least one will neither have the disease nor be a carrier?
```
ans = P(NNN) +
      P(NND) + P(NDN) + P(DNN) +
      P(NNC) + P(NCN) + P(CNN) +
      P(NCD) + P(NDC) + P(NCC) + P(NDD) +
      P(CND) + P(DNC) + P(CNC) + P(DND) +
      P(CDN) + P(DCN) + P(CCN) + P(DDN)
ans = (0.25 * 0.25 * 0.25) +
      3 * (0.25 * 0.25 * 0.25) + 
      3 * (0.25 * 0.25 * 0.50) + 
      3 * ((0.25 * 0.25 * 0.50) + (0.25 * 0.25 * 0.50) + (0.25 * 0.50 * 0.50) + (0.25 * 0.25 * 0.25))
ans = 0.578125
```

````python
prob = [0.25, 0.50, 0.25] # [Disease, Carrier, Neither carrier nor disease]
options = ['D', 'C', 'N'] # [Disease, Carrier, Neither carrier nor disease]
count = 0
simulations = 0
x = []
y = []

while simulations < 10000:
    child1 = np.random.choice(options, p=prob)
    child2 = np.random.choice(options, p=prob)
    child3 = np.random.choice(options, p=prob)
    if child1 == 'N' or child2 == 'N' or child3 == 'N':
        count += 1
    simulations += 1
    if simulations != 0 and (simulations % 10) == 0:
            y.append(float(count) / float(simulations))
            x.append(simulations)

fig2 = plot_probs(x, y)
iplot(fig2)

prob = 100*float(count/simulations)
print('Probablity at lease one will neither have the disease nor be a carrier  %.2f%%' % prob)
````
